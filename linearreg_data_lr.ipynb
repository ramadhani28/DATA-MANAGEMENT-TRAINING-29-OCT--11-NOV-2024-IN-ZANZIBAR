```python
#LINEAR REGRESSION IMPLEMENTATION
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
data=pd.read_csv('data_for_lr.csv')
data.isna().sum()
data.shape
data.to_csv('data_for_lr_copy.csv')
```

```python
# Mean imputation(fill missing values with past observation)
data1 = data.fillna(data.y.mean())
data1.shape
data1.isnull().sum()
# for this exercise please use dropna
data2 = data.dropna()
data2.isnull().sum()
```

```python
#graphical representation
figure, axis = plt.subplots(2,2)
plt.scatter(data2['x'], data2['y'])
plt.show()
```

```python
X=data2['x']
Y=data2['y']
X_train,X_test,Y_train,Y_test = train_test_split(np.array(X).reshape(-1,1),Y,test_size=0.2,random_state=2)
X
# linear regression
reg = LinearRegression()
reg.fit(X_train, Y_train)
print(reg.intercept_) # this is beta zero
print(reg.coef_) # this is beta one(1)
#prediction
Y_pred = reg.predict(X_test)
#print(Y_pred,Y_test)
r2_score(Y_pred,Y_test)
u=reg.predict(np.array(102).reshape(-1,1))
print(u)
```
