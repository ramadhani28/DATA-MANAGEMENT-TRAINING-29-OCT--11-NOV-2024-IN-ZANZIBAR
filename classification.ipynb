```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# prediction of chance of admission(COA)
admissiondataset=pd.read_csv('Admission_Predict.csv')
#admissiondataset.columns
#admissiondataset.info()
admissiondataset.head()
```

```python
# threshold value for target variable and change value of COA with category value ie greater than 75 1 less than 75 0
admissiondataset['COA'] = np.where(admissiondataset['COA']>.75,1,0)
#admissiondataset['COA'].value_counts()
```

```python
Y = admissiondataset['COA']
X = admissiondataset.drop(['Serial No.','COA'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=.2,random_state=1)
logmodel = LogisticRegression(random_state=1)
logmodel.fit(X_Train,Y_Train)
Y_Predict = logmodel.predict(X_Test)
from sklearn.metrics import accuracy_score
accuracy_score(Y_Test,Y_Predict)
#Save your trained model
#import joblib
#joblib.dump(logmodel, 'model.pkl')
```

```python
#create a data frme to show X_test,Y_test and Y predict 
Analysis=X_Test.copy()
Analysis['Y_Test']=Y_Test
Analysis['Y_Predict']=Y_Predict
Analysis
```

```python
# Identify wrong predictions
wrong_predictions = Analysis[Analysis['Y_Test'] != Analysis['Y_Predict']]

print(wrong_predictions)
```
