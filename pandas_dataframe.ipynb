```python
#loading excel file into dataframe
import pandas as pd
df = pd.read_excel('salesdump.xlsx')
#for absolute path please use df = pd.read_excel(r'C:\Users\vaina\Downloads\salesdump.xlsx')
print(df.head(2))
```

```python
#print first row using loc function
#with loc data can be fetched using some condition also
print(df.loc[0])
print('\n')
print(df.loc[0][4])# this is first row fifth column
print('\n')
print(df.loc[df.Quantity>=10])
# slicing can also be used
print(df.loc[2:4]['Product_Type'])
print('\n')
print(df.loc[(df.Customer_ID==1000) & (df.Quantity==7)]['Quantity'])
print('\n')
print(df.loc[(df.Customer_ID==1000) | (df.Quantity==7)]['Quantity'])
print('\n')
# loc can be used for data update
df.loc[(df.Customer_ID==1000) & (df.Quantity==7),['Quantity']] = 18
print(df.loc[df.Customer_ID==1000]['Quantity'])
```

```python
#demo to display columns
print(df.loc[0][['Customer_ID','Quantity']]) ##SELECT Customer_ID,Quantity from files WHERE index = 0;
print('\n')
print(df[['Customer_ID','Quantity']]) ##SELECT Customer_ID,Quantity from files ;
```

```python
# display rows
print(df.iloc[0:2])
#selective rows can also be displayed
print('\n')
print(df[['Product_Type', 'Unit_Price']].iloc[[2, 10]])
print('\n')
#selective rows and columns
print(df['Product_Type'].iloc[0:5])
print('\n')
print(df[['Product_Type','Unit_Price']].iloc[2:10])
```

```python
#adding new record in dataframe
import pandas as pd
data = {'Name':['Manish', 'Jha', 'Himanshu'],
       'Empcode':[10,11,12]}
# Create Dataframe
df=pd.DataFrame(data)
#print the output
print(df, '\n')
newRow = [{'Name':'Khan','Empcode':20}, {'Name':'slamat','Empcode':21}]
dfNew = pd.DataFrame(newRow)
df2=pd.concat([df,dfNew])
df2=pd.concat([df,dfNew]).reset_index(drop=True)
print(df2)
```
