```python
import numpy as np
import pandas as pd
from scipy import stats
df = pd.read_csv('quikr_car.csv')
#df.head()
#df.info()
#df.isna().sum()
#df['Price'].unique()
df=df[df['Price']!='Ask For Price']
#df['Price'].unique()
df.isnull().sum()
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# remove , from Price
df['Price'] = df['Price'].str.replace(',','').astype(int)
df.info()
```

```python
# change for year too
#df=df[df['year'].str.isnumeric()]
df['year'] = df['year'].astype(int)
df.info()
```

```python
# cahnge for kms_driven
df['kms_driven'].unique()
#df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',','').astype(int)
df['kms_driven'].unique()
df.info()
```

```python
#here we deal with categorization
df['fuel_type'].value_counts()
df.replace({'fuel_type':{'Petrol':0,'Diesel':1,'LPG':2}},inplace=True)
df.fuel_type.unique()
```

```python
df.isnull().sum()
df.info()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#model creation
X=df.drop(['name','company','Price'],axis=1)
Y=df['Price']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.2,random_state=2)
X
```

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

```python
lm.coef_
lm.intercept_
#prediction
Y_Predict=lm.predict(X_test)
data={"year":[2004],"kms_driven":[9000],"fuel_type":[1]}
datatotest=pd.DataFrame(data)
lm.predict(datatotest)
```

```python
#Accurcy
accuracy=r2_score(Y_test,Y_Predict)
print("Accuracy of the model is :",accuracy)
```
