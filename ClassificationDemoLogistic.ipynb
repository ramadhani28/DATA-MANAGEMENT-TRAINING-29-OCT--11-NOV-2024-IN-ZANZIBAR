```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
admissiondataset=pd.read_csv('AI-MLData/Admission_Predict.csv')
#admissiondataset.columns
#admissiondataset.info()
admissiondataset.head()
```

```python
admissiondataset['COA']=np.where(admissiondataset['COA']>.75,1,0)
```

```python
#admissiondataset['COA'].value_counts()
```

```python
Y=admissiondataset['COA']
X=admissiondataset.drop(['Serial No.','COA'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=.2,
                                               random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression(random_state=1)
logmodel.fit(X_Train,Y_Train)
Y_Predict=logmodel.predict(X_Test)
from sklearn.metrics import accuracy_score
accuracy_score(Y_Test,Y_Predict)

# Save your trained model
#import joblib
#joblib.dump(logmodel, 'model.pkl')
```

```python
Analysis=X_Test.copy()
Analysis["Y_Test"]=Y_Test
Analysis["Y_Predict"]=Y_Predict
Analysis
```
