# The following code is used to plot graphs. First it load data of tripduration and temperature.

```python
riderDF = pd.read_excel("data_rider_v2.xlsx", usecols='K,L') 
gbTemp = riderDF.groupby("temperature").count()
print(gbTemp)
```

# Scatter Plot using Dataframe

```python
import matplotlib.pyplot as plt
riderDF.plot(kind='scatter',
        x='tripduration',
        y='temperature',
        color='blue')

# set the title
plt.title('ScatterPlot-Rider')
# show the plot
plt.show()
```

# Line using Dataframe

```python
import matplotlib.pyplot as plt
from webcolors import rgb_to_name
riderDF.plot(kind='line',
        x='tripduration',
        y='temperature',
        color=rgb_to_name((123, 104, 238), spec='css3'))
# set the title
plt.title('Line-Rider')
# show the plot
plt.show()
```

# Histogram using Dataframe, Histogram shows the frequency of data. The bin attribute is used to divide the data. By default it is 10.

```python
import matplotlib.pyplot as plt
dfhisto = riderDF.loc[:, 'temperature']
dfhisto.plot(kind='hist',bins=15, edgecolor ='black',
        color=rgb_to_name((123, 104, 238), spec='css3'))
# set the title
plt.title('Histogram-Rider')
plt.xlabel('Temperature')
plt.ylabel('Number of Trips')
# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
partDF = riderDF.head(30)
partDF.plot(kind='bar',
        x='temperature',
        y='tripduration',
        color='red')
# set the title
plt.title('Bar-Rider')
plt.xlabel('Temperature')
plt.ylabel('Number of Trips')
# show the plot
plt.show()
```

# The following code group the data by temperature and count the number of trips. When groupby function is used with dataframe, it returns object. 
# The object needs to be converted to dataframe again. 
# The method, .reset_index() converts object to dataframe. And then converted dataframe is used to draw the bar.

# For practice, only first 200 data is used

```python
import matplotlib.pyplot as plt
partDF = riderDF.head(200)
obj = pd.DataFrame(partDF.groupby("temperature")['tripduration'].count())
newDF=obj.reset_index()
print(newDF)
newDF.plot(kind='bar',
        x='temperature',
        y='tripduration',
        color='blue')
# set the title
plt.title('Bar-Rider')
plt.xlabel('Temperature')
plt.ylabel('Number of Trips')
# show the plot
plt.show()
```

# we can also use as_index =False in group by to return DF as shown below

```python
import matplotlib.pyplot as plt
partDF = riderDF.head(200)
obj = pd.DataFrame(partDF.groupby("temperature",as_index=False)['tripduration'].count())
#newDF=obj.reset_index()
print(newDF)
obj.plot(kind='bar',
        x='temperature',
        y='tripduration',
        color='blue')
# set the title
plt.title('Bar-Rider')
plt.xlabel('Temperature')
plt.ylabel('Number of Trips')
# show the plot
plt.show()
```

# The following code calculates the number of trips according to range.
# for practise purpose, the following ranges of temperature are defined
# 1- Less then zero
# 2- Zero to 20
# 3- 20 to 40
# 4- 40 to 60
# 5- 60 to 80
# 6- 80 to 100
# First, calculation is done individually and and then dictionary is created. From dictionary, dataframe is created. 

```python
riderDF.columns
ata = [1, 2, 3, 4, 5]

# Convert to DataFrame
temp_lt0 = riderDF.loc[(riderDF['temperature'] <= 0)].shape[0]
temp_0_20 = riderDF.loc[(riderDF['temperature'] > 0) & (riderDF['temperature'] <= 20)].shape[0]
temp_20_40 = riderDF.loc[(riderDF['temperature'] > 20) & (riderDF['temperature'] <= 40)].shape[0]
temp_40_60 = riderDF.loc[(riderDF['temperature'] > 40) & (riderDF['temperature'] <= 60)].shape[0]
temp_60_80 = riderDF.loc[(riderDF['temperature'] > 60) & (riderDF['temperature'] <= 80)].shape[0]
temp_80_100 = riderDF.loc[(riderDF['temperature'] > 80) & (riderDF['temperature'] <= 100)].shape[0]
freqcount =[temp_lt0, temp_0_20,temp_20_40, temp_40_60, temp_60_80,temp_80_100]
freqrange=['lt 0','0-20','20-40','40-60','60-80','80-100']
freqname=['Range', 'Count']
dict = {'Range': freqrange, 'Count': freqcount} 
newDF = pd.DataFrame(dict)
print(newDF)
```

# The following code  draws bar chart based on newly created Dataframe

```python
newDF.plot(kind='bar', 
        x='Range',
        y='Count',
        color='blue')
plt.legend(['Number of Trips'])
plt.title('Bar-Rider')
plt.xlabel('Range')
plt.ylabel('Trip Count')
# show the plot
plt.show()
```
